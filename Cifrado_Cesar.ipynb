### Principios del Cifrado César

El Cifrado César es uno de los métodos de cifrado más simples y antiguos. Es un tipo de cifrado por sustitución en el que cada letra del texto original es reemplazada por una letra que se encuentra un número fijo de posiciones más adelante en el alfabeto. Este número fijo es lo que se conoce como la **clave** del cifrado.

#### Desplazamiento de Letras

El desplazamiento de letras es el corazón del Cifrado César. Consiste en mover cada letra un cierto número de posiciones en el alfabeto. Por ejemplo:

*   Si la clave (desplazamiento) es **3**:
    *   'A' se convierte en 'D'
    *   'B' se convierte en 'E'
    *   'X' se convierte en 'A' (envuelve el alfabeto)
    *   'Y' se convierte en 'B' (envuelve el alfabeto)
    *   'Z' se convierte en 'C' (envuelve el alfabeto)

#### Concepto de la Clave

La **clave** en el cifrado César es un número entero que determina cuántas posiciones se desplaza cada letra. Es esencial que tanto el emisor como el receptor conozcan esta clave para poder cifrar y descifrar el mensaje, respectivamente. Sin la clave, el descifrado se vuelve una tarea de "fuerza bruta" probando cada posible desplazamiento (hay 25 posibles claves si se usa un alfabeto de 26 letras).

#### Envoltura del Alfabeto

Un aspecto crucial del cifrado César es cómo maneja las letras al final del alfabeto. Cuando una letra cifrada excede la 'Z' (o la última letra del alfabeto utilizado), el conteo "envuelve" o "rota" al principio del alfabeto. Por ejemplo, si el desplazamiento es 3 y la letra original es 'X', la 'X' se movería 3 posiciones: 'Y', 'Z', y luego 'A'.

#### Consideraciones Adicionales

Tradicionalmente, el cifrado César:

*   No distingue entre mayúsculas y minúsculas (por ejemplo, 'a' y 'A' se cifran de la misma manera).
*   Ignora caracteres no alfabéticos, como números, signos de puntuación y espacios, manteniéndolos sin cambios en el texto cifrado.

```python
def caesar_cipher_encrypt(message, shift):
    encrypted_message = ""
    for char in message:
        if 'A' <= char <= 'Z':
            encrypted_message += chr(((ord(char) - ord('A') + shift) % 26) + ord('A'))
        elif 'a' <= char <= 'z':
            encrypted_message += chr(((ord(char) - ord('a') + shift) % 26) + ord('a'))
        else:
            encrypted_message += char
    return encrypted_message


message_to_encrypt = "hola buenas tardes"
shift_key = 12
encrypted_text = caesar_cipher_encrypt(message_to_encrypt, shift_key)
print(f"mensaje Original : {message_to_encrypt}")
print(f"clave: {shift_key}")
print(f"mensaje encriptado: {encrypted_text}")
```

```python
import collections

def calculate_letter_frequencies(text):
    letter_counts = collections.defaultdict(int)
    total_alphabetic_letters = 0

    for char in text:
        upper_char = char.upper()
        if ('A' <= upper_char <= 'Z') or (upper_char == 'Ñ'):
            letter_counts[upper_char] += 1
            total_alphabetic_letters += 1

    letter_frequencies = {}
    if total_alphabetic_letters > 0:
        for letter, count in letter_counts.items():
            letter_frequencies[letter] = count / total_alphabetic_letters

    return letter_frequencies

sample_message = "Hola Mundo"
frequencies = calculate_letter_frequencies(sample_message)

print(f"mensaje original : {sample_message}")
print("frecuencia:")
for letter, freq in sorted(frequencies.items()):
    print(f"  '{letter}': {freq:.4f}")


sum_of_frequencies = sum(frequencies.values())
print(f"\n suma de frecuencia: {sum_of_frequencies:.4f}")


assert abs(sum_of_frequencies - 1.0) < 1e-9 or sum_of_frequencies == 0.0, "Frequencies do not sum to 1.0"
print("Frequencies sum check passed.")
```

```python
def caesar_bruteforce_decrypt(encrypted_message):
    print(f"mensaje cifrado: '{encrypted_message}'\n")
    for shift in range(1, 26):
        decrypted_text = caesar_cipher_encrypt(encrypted_message, 26 - shift)
        print(f"desplazamiento y resultador {shift}: {decrypted_text}")

encrypted_message_for_bruteforce = "Welcome to Home"
caesar_bruteforce_decrypt(encrypted_message_for_bruteforce)
```

```python
spanish_frequencies = {
    'E': 0.125,
    'A': 0.117,
    'O': 0.093,
    'L': 0.080,
    'S': 0.079,
    'N': 0.072,
    'D': 0.055,
    'R': 0.055,
    'U': 0.047,
    'I': 0.046,
    'T': 0.046,
    'C': 0.038,
    'P': 0.031,
    'M': 0.030,
    'Y': 0.019,
    'Q': 0.015,
    'B': 0.012,
    'H': 0.012,
    'G': 0.010,
    'F': 0.007,
    'V': 0.007,
    'J': 0.005,
    'Ñ': 0.003,
    'Z': 0.003,
    'X': 0.002,
    'K': 0.001,
    'W': 0.001
}
```

```python
def calculate_chi_squared(observed_frequencies, expected_frequencies):
    chi_squared_distance = 0
    for char in expected_frequencies:
        expected_freq = expected_frequencies.get(char, 0)
        observed_freq = observed_frequencies.get(char, 0)

        if expected_freq > 0:
            chi_squared_distance += ((observed_freq - expected_freq) ** 2) / expected_freq
    return chi_squared_distance

sample_message_frequencies = calculate_letter_frequencies("Hola Mundo")
chi_distance = calculate_chi_squared(sample_message_frequencies, spanish_frequencies)
print(f"Chi-squared distance for 'Hola Mundo': {chi_distance:.4f}")

hello_world_freq = calculate_letter_frequencies("HELLO WORLD")
chi_hello_world = calculate_chi_squared(hello_world_freq, spanish_frequencies)
print(f"Chi-squared distance for 'HELLO WORLD': {chi_hello_world:.4f}")

random_frequencies = {'Z': 0.5, 'Q': 0.3, 'X': 0.2}
chi_random = calculate_chi_squared(random_frequencies, spanish_frequencies)
print(f"Chi-squared distance for random frequencies: {chi_random:.4f}")
```

```python
import math

def caesar_frequency_decrypt(encrypted_message):
    best_score = math.inf
    best_decrypted_text = ""
    best_shift = 0

    for shift in range(1, 26):
        decrypted_text_attempt = caesar_cipher_encrypt(encrypted_message, 26 - shift)
        current_frequencies = calculate_letter_frequencies(decrypted_text_attempt)
        current_score = calculate_chi_squared(current_frequencies, spanish_frequencies)

        if current_score < best_score:
            best_score = current_score
            best_decrypted_text = decrypted_text_attempt
            best_shift = shift

    return best_decrypted_text, best_shift

encrypted_message_ai = "KHOOR ZRUOG"
decrypted_ai_text, detected_shift = caesar_frequency_decrypt(encrypted_message_ai)

print(f"\nmensaje encriptado: {encrypted_message_ai}")
print(f"Best decrypted message (AI): {decrypted_ai_text}")
print(f"saltos detectados: {detected_shift}")

encrypted_message_ai_2 = "MTQF RZSIT"
decrypted_ai_text_2, detected_shift_2 = caesar_frequency_decrypt(encrypted_message_ai_2)

print(f"\nmensaje encriptado: {encrypted_message_ai_2}")
print(f"Best decrypted message (AI): {decrypted_ai_text_2}")
print(f"saltos detectados: {detected_shift_2}")
```

```python
print("\n--- prueba  AI ---")

test_cases = [
    {"original_message": "HOLA MUNDO ESTO ES UNA PRUEBA", "shift": 5},
    {"original_message": "ENCRIPTACION ES DIVERTIDA", "shift": 10},
    {"original_message": "LA COMPUTADORA NO PUEDE PENSAR SOLO PUEDE EJECUTAR INSTRUCCIONES", "shift": 3},
    {"original_message": "EL SOL BRILLA FUERTE HOY", "shift": 15},
    {"original_message": "CINCO MONOS SALTANDO EN LA CAMA", "shift": 20},
    {"original_message": "LA LENGUA ES UN ORGANO MUSCULOSO", "shift": 7},
    {"original_message": "LA CIBERSEGURIDAD ES IMPORTANTE", "shift": 12},
    {"original_message": "EL NIÑO COME Y JUEGA", "shift": 4} # Includes 'Ñ'
]

results = []

for i, test_case in enumerate(test_cases):
    original_msg = test_case["original_message"]
    shift_key = test_case["shift"]

    # Encrypt the message
    encrypted_msg = caesar_cipher_encrypt(original_msg, shift_key)

    # Decrypt using the AI
    decrypted_ai_text, detected_shift = caesar_frequency_decrypt(encrypted_msg)

    print(f"\n--- Test Case {i+1} ---")
    print(f"Original Message: {original_msg}")
    print(f"Encryption Shift: {shift_key}")
    print(f"Encrypted Message: {encrypted_msg}")
    print(f"AI Decrypted Message: {decrypted_ai_text}")
    print(f"AI Detected Shift: {detected_shift}")

    is_correct_message = (original_msg == decrypted_ai_text)
    is_correct_shift = (shift_key == detected_shift)
    print(f"AI message correct: {is_correct_message}")
    print(f"AI shift correct: {is_correct_shift}")
    results.append({
        "original": original_msg,
        "encrypted": encrypted_msg,
        "encryption_shift": shift_key,
        "ai_decrypted": decrypted_ai_text,
        "ai_detected_shift": detected_shift,
        "message_correct": is_correct_message,
        "shift_correct": is_correct_shift
    })

print("\n--- Summary of AI Decryption Tests ---")
successful_decryptions = 0
successful_shifts = 0
for res in results:
    if res["message_correct"]:
        successful_decryptions += 1
    if res["shift_correct"]:
        successful_shifts += 1
    print(f"- Original: '{res['original']}' -> AI Decrypted: '{res['ai_decrypted']}' (Shift {res['encryption_shift']} -> Detected {res['ai_detected_shift']}) - Message Correct: {res['message_correct']}, Shift Correct: {res['shift_correct']}")

print(f"\nTotal test cases: {len(test_cases)}")
print(f"Messages correctly decrypted: {successful_decryptions}/{len(test_cases)}")
print(f"Shifts correctly identified: {successful_shifts}/{len(test_cases)}")
```
